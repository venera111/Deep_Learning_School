* `list(range(start, end[, step]))` - получить последовательность (список) целых чисел, начинающуюся со `start`, заканчивающуюся в `end-1` и шагом `step`

### enumerate, zip

```python
first = 'a b c d e f g'.split(' ') #list
second = '1 2 3 4 '.split(' ') #list
print(first, second)
zip(first, second)
```

Если мы передаем объекты методу zip, он итерируется по всем элементам переданных объектов одновременно, и возвращает i-е элементы всех объектов.

```python
list(zip(first, second))
```

Как только элементы заканчиваются в одном из list, то итерация прекращается. Можно передавать бесконечное количество аргументов в zip

```python
list(zip(first, second * 3, first, second * 2)) # размножим second так, что он больше чем first
```

zip возвращает итератор, чтобы явно им воспользоваться, можно преобразовать его к list, либо использовать iter(), который уже определен в zip() по умолчанию. Используем оператор next() для итерации:

Цикл for создает автоматически итератор неявно. И на каждой итерации зовет метод next() до тех пор, пока не придет StopIteration. 

```python
a = zip(first, second)
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a)) # здесь вызовется StopIterator так как список second закончился
```

Иногда мы итерируемся по объекту и хотим знать его индекс, чтобы не использовать цикл for и метод index. Так как это заставляет метод index заново проходиться по всем элементам (время O(n)). enumerate возвращает индекс и сам объект в виде кортежа tuple

```python
methods = dir(__builtin__)

for num, method in enumerate(methods):
    if num % 5 == 0:
        print(num, method)
```

```python
l = ['a', 'b', 'a']
c = [0, 0, 0]

for idx, person in enumerate(l):
    if person == 'a':
        c[idx] += 10000


print(list(zip(l, c))) # соединим итоговые списки
```

Выход enumerate можем явно преобразовать к list, получим все пары сразу в list. Но тогда все эти пары (индекс, значение) будут храниться в памяти. И если мы заходим проитерироваться до 1 миллиарда, то лучше это делать при помощт цикла и не хранить эти данные в оперативной памяти.

```python
enum = enumerate(first)
list(enum)
```

```python
first
```

Пока сравниваем два списка `list(enum) == list(zip_style)`, мы обращаемся к обоим итераторам и проходим до конца. Если после этой строки вызвать распечатку zip_style, то будет пустой список, а если потом вызвать `next(zip_style)`, то выйдет `StopIteration`, так как мы в конце.

```python
enum = enumerate(first) # нумеруем значения списка first
zip_style = zip(range(0, len(first)), first) # склеим в пары индексы и сами значения списка first

print(list(enum) == list(zip_style)) # в итоге списки пар равны
print(list(zip_style))
# next(zip_style)
```

Для повторного использования итератора нужно было либо преобразовать его к листу, либо копировать при помощи  метода `itertools.tee`

```python
from itertools import tee

values = ['a', 'b', 'c', 'd']
enum = enumerate(values)
zip_style = zip(range(0, len(values)), values)
for_comparison, for_printing = tee(zip_style)
print(list(enum) == list(for_comparison))

list(for_printing)
```

### list comprehensions

Мы можем создавать списки большой длины или с условием. Внутри квадратных скобок мы пишем целый цикл: "положи все элементы (икс) которые принимают значения по итератору range от 1 до 6 (не вкл)"

```python
a = [x for x in range(1, 6)]
a
```

Можно положить в список результат функции:

```python
def f(x):
	return x ** 2
```

```python
b = [f(x) for x in range(1, 10)]
c = [x ** 2 for x in range(1, 10)] # определяем функцию явно
print(b, '==', c)
print(b == c)
```

В list comprehensions помимо циклом можно использовать условную логику. Мы итерируемся по всем символам строки `'apple'`, затем применяем условия if/else и создаем список с теми символами, которые содержатся в строке `'aeiou'`, заменяя остальные звездочками:

```python
[x if x in 'aeiou' else '*' for x in 'apple']
```

```python
l = []
for i in range(3):
	for x in [i, i + 1]:
		l.append(str(x))
	print(l)
l
```

Этот код можно записать намного проще:

```python
l = [str(x) for i in range(3) for x in [i, i + 1]]
l
```

```python
def make_coffee(size, sugar_dose=3, **kwargs):
	print(kwargs)
	print(kwargs['name'])
	if sugar_dose > 5:
		return 'Too much sugar! Be careful! :('
	else:
		return f'Done: cup of {size} ml size; amount of sugar = {sugar_dose}'
```

```python
make_coffee(100, 6)
```

Распечатаем в функции make_coffee аргументы kwargs `print(kwargs)` и после распечатаем значение аргумента kwargs по  именованию `print(kwargs['name'])`. Заранее раскомментируем эти строки в определении функции.

```python
make_coffee(120, 5, name='Ilya', gender='male')  # передаем именованные арг-ты в kwargs
```

### map, reduce, filter

* `map(func, iterables)` - выполняет преобразование func над элементами iterables и возвращает **новый** `iterable`:

```python
words = dir(list)
type(words)
```

Первым аргументом в map мы передаем функцию, которую будем применять ко второму аргументу, например к списку. И функция применяется ко всем элементам списка (к элементам итерируемого объекта)

```python
letter_counts = list(map(lambda x: len(x), words)) # список длин названий методов list
print(letter_counts)
print()
print(words)
```

map принимает не ограниченное число аргуметов, он применяет итерацию похожую на zip. Например, есть 3 списка и мы складываем поэлементно эти списки:

```python
l1 = [1,2,3,4]
l2 = [11, 12, 13, 14, 15]
l3 = [101, 102, 103]

triple_sum = list(map(lambda x, y, z: (x, y, z), l1, l2, l3)) # просто выводим элементы итеративно
print(triple_sum)
```

```python
k1 = [10, 20, 30, 40]
k2 = [100, 200, 300]
k3 = [1000, 2000, 3000, 4000]

result = map(lambda x, y, z: x + y + z, k1, k2, k3) # складываем поэлементно списки и получаем объект map
print(result)
result = list(result) # приводим map к list (преобразование кортежа в список)
print(result)
```

* `reduce(func, iterables)` - производит вычисление с элементами последовательности, результатом которого является **одно значение**:

```python
from functools import reduce
```

Посчитаем сумму всех длин с помощью `reduce`. При первом вызове lambda берутся первые 2 аргумента, они складываются. На следующей итерации эта сумма становится `x`. И так проходим до конца, пока не посчитаем всю сумму:

```python
sum_of_counts = reduce(lambda x, y: x + y, letter_counts)
print(sum_of_counts)
```

* `filter(predicate, iterable)` - оставляет только те элементы, для которых **верен** предикат (функция, возвращающая bool):

```python
mixed = ['мак', 'просо', 'мак', 'мак', 'просо', 'мак', 'просо', 'просо', 'просо', 'мак']
only_mac = list(filter(lambda x: x == 'мак', mixed))	# lambda x: x == 'мак' выдает True/False
print(only_mac)

[x for x in mixed if x == 'мак']
```

Мы оставили в списке те элементы, которые соответствовали True в функции-предикате. Также можно сделать такой же список с помощью list comp.
Из-за низкой читабельности filter map reduce чаще используют list comp, так как он читаемее.

### classes

У списка есть `__str__` метод для преобразования его в строку. И этот метод вызывается в классе ниже:

```python
f"list: {[1, 2, 3]}"
dir(list)
```

Класс создается с помощью метода __init__, который принимает на вход аргументы. Можем задать значения по умолчанию:

```python
class Human:
    def __init__(self, name='', age=None, deep_learning_specialist=False):
        self.name = name
        self.age = age
        self.deep_learning_specialist = deep_learning_specialist
    
    def set_age(self, age): # задает возраст конкретному объекту класса
        self.age = age
    
    def get_age(self): # возвращает возраст объекта
        return self.age

    def test(self):
        print("Aaa")
    
    def __str__(self): # метод преобразования к строке (стандартный метод python для list)
        return f'Name: {self.name}\nAge: {self.age} \
                \nIs deep learning specialist: {self.deep_learning_specialist}' # код становится короче


class DLSchoolStudent(Human):
    def __init__(self, name='', age=None, deep_learning_specialist=True):
        super().__init__(name, age) # изначально создается Human, это конструктор для всех базовых классов от которого наследовались
        self.total_grade = None
        self.deep_learning_specialist = deep_learning_specialist

    def test(self): # существующие методы перезаписываются
        print("BBB")
```

Создаем объект класса `Human` и передаем ему значения:

```python
human = Human('Person', 17)
print(human)
human.get_age() # вызываем метод класса
```

Наследование. Создаем класс и в скобках пишем, от какого класса наследовать все базовые свойства. 

```python
student = DLSchoolStudent('Good Person', 18)
print(student)
```

```python
student.test()
```

метод `open` открывает файл, находящийся по пути и с режимом. У нас создается файловый дескриптор в C, затем нам обязательно нужно fd закрыть.

```python
file = open(path, mode='r') # только на чтение
print([line for line in file][0])
file.close()
```

| Режим | Обозначение |
|-------|-------------|
| **'r'**  | Открытие на **чтение** (является значением по умолчанию) |
| **'rb'** | Открытие на **чтение**, в предположении, что будут считываться **байты** |
| **'w'** | Открытие на **запись**, содержимое файла удаляется. Если файла не существует, создается новый |
| **'wb**' | Открытие на **запись байтов**, содержимое файла удаляется. Если файла не существует, создается новый |
| **'a'** | Открытие на **дозапись**, информация добавляется **в конец файла** |
| **'r+'** | Открыть файл на **чтение И запись**. Если файла нет, **новый НЕ создаётся** |
| **'a+'** | Открыть файл на **чтение И запись в конец файла**. Если файла нет, **новый создаётся** |
| **'t'** | Открытие файла **как текстового** (по умолчанию) |

С помощью контекстного менеджера with открываем файл на чтение и печатаем все строки. И как только код выполняется, файловый дескриптор закрывается и нам за ним следить не нужно.

```python
with open(path, mode='r') as test_file:
    for line in test_file:
        print(line)
```

### collections

* `defaultdict()` - класс словаря, у которого есть значение по умолчанию - порой очень пригождается:

```python
from collections import defaultdict
```

```python
d = defaultdict(int) # создает словарь с дефолтным значением 0 | "" | []
print(d.get('key', 0)) # возвращается дефолтное значение

d['key'] = 5
print(d['key'])  # 5
```

Мы можем передать функцию, которая означает значение по умолчанию. Если включа `key` нет, то возвращается по умолчанию, а если есть ключ, то присваиваем ему значение другое:

```python
d = defaultdict(lambda: 'empty')
print(d['key'])

d['key'] = 'full'
print(d['key'])
```

Дефолтным значением может быть список. И мы можем добавлять по ключу целый список:

```python
d = defaultdict(list)
print(d)
d['list1'].append(100) # добавляем новый элемент в список по ключу словаря
d['list1'].append(200) # добавляем еще один новый элемент в список по ключу словаря
print(d)
print(d['list1']) # печатаем содержимое по ключу словаря (список с друмя добавленными элементами)
```

Этот метод очень удобен, так как нам не нужно проверять, существует ли список по данному ключу или нет. А если делать через `dict()`, то проверка необходима:

```python
d = dict()
print(d)
if 'list1' in d.keys(): # если ключ есть то элементы добавляем
    d['list1'].append(100)
    d['list1'].append(200)
else: # если ключа нет, то по ключу записываем пустой список
    d['list1'] = []
print(d)
print(d['list1'])
```

* `Counter()` - класс словаря, предназначенного для счётчиков. По сути, `== defaultdict(int)`: Позволяет считать элементы в объекте

```python
from collections import Counter
```

Посчитаем все символы из названий метода `__builtin__`

```python
counter = Counter()
for word in dir(__builtin__):
    for letter in word:
        counter[letter] += 1  # или .update(value)
print(counter)
```

```python
print(counter['P'])
```

Counter() позволяет создать словарь, где ключом являются символы, а значениями - их частота. С ним можно работать как со славарем, получая значения по ключу.
